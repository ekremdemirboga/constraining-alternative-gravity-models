```python
import numpy as np
import matplotlib.pyplot as plt
import glob
import os
from matplotlib import cm
```

```python
from sys import api_version


def read_data(filename):
    r=[]
    mu=[]
    psi = []
    phi=[]
    pressure = []
    rho = []
    indepres = []
    res1 = []
    res2 = []
    res3 = []
    res4 = []


    datafile=open(filename,'r')
    lines = datafile.readlines()[1:]
    for line in lines:
        (a,b,c,d,e,f,g,g1,g2,g3,g4)=line.split()
        r.append(float(a))
        mu.append(float(b))
        psi.append(float(c))
        phi.append(float(d))
        pressure.append(float(e))
        rho.append(float(f))
        indepres.append(float(g))
        res1.append(float(g1))
        res2.append(float(g2))
        res3.append(float(g3))
        res4.append(float(g4))


    datafile.close()
    return np.array(r),np.array(mu),np.array(psi),np.array(phi),np.array(pressure),np.array(rho),np.array(indepres),np.array(res1),np.array(res2),np.array(res3),np.array(res4)
```

```python
numberOfFiles = 12
txt_files = sorted(glob.glob('star2/*.txt'), key=os.path.getmtime)
colors = cm.rainbow(np.linspace(0, 1.2, numberOfFiles))
h = 1 # (order of convergence )**
```

increasing number indicates increasing mesh points, starting from n=64 to n=131072. The following is the solution for beta = +200 (not code units) and mass_st = 0.

```python
##Independent Residuals##
r,_,_,_,_,_,_,_,_,_,_= read_data(txt_files[0])
plt.figure()
fig = plt.gcf()
fig.set_size_inches(8, 4.5)
plt.title("residuals")
i=0
for filename in txt_files[0:]:
    _,_,_,_,_,_,indepres,resi1,res2,res3,res4= read_data(filename)
    plt.loglog(r, np.abs(indepres[::2**(i)])*h**i,color=colors[i],label=i)
    i+=1
plt.grid()
plt.legend(loc='right')
# plt.ylim(-8e-1, 8e-1)


i=0
plt.figure()
plt.title("mu equation")
fig = plt.gcf()
fig.set_size_inches(8, 4.5)
for filename in txt_files[0:]:
    _,_,_,_,_,_,indepres,resi1,res2,res3,res4= read_data(filename)
    plt.loglog(r, np.abs(resi1[::2**(i)])*h**i,color=colors[i],label=i)
    i+=1
plt.grid()
plt.legend(loc='right')
plt.ylim(1e-15, 1e-1)


i=0
plt.figure()
plt.title("psi equation")
fig = plt.gcf()
fig.set_size_inches(8, 4.5)
for filename in txt_files[0:]:
    _,_,_,_,_,_,indepres,resi1,res2,res3,res4= read_data(filename)
    plt.loglog(r, np.abs(res2[::2**(i)])*h**i,color=colors[i],label=i)
    i+=1
plt.grid()
plt.legend(loc='right')
# plt.ylim(-8e-1, 8e-1)


i=0
plt.figure()
plt.title("q equation")
fig = plt.gcf()
fig.set_size_inches(8, 4.5)
for filename in txt_files[0:]:
    _,_,_,_,_,_,indepres,resi1,res2,res3,res4= read_data(filename)
    plt.loglog(r, np.abs(res3[::2**(i)])*h**i,color=colors[i],label=i)
    i+=1
plt.grid()
plt.legend(loc='right')
# plt.ylim(-8e-1, 8e-1)


i=0
plt.figure()
plt.title("phi equation")
fig = plt.gcf()
fig.set_size_inches(8, 4.5)
for filename in txt_files[0:]:
    _,_,_,_,_,_,indepres,resi1,res2,res3,res4= read_data(filename)
    plt.loglog(r, np.abs(res4[::2**(i)])*h**i,color=colors[i],label=i)
    i+=1
plt.grid()
plt.legend(loc='right')
# plt.ylim(-8e-1, 8e-1)
```
