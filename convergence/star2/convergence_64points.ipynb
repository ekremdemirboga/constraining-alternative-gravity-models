```python
import numpy as np
import matplotlib.pyplot as plt
import glob
import os
from matplotlib import cm
```

```python
from sys import api_version


def read_data(filename):
    r=[]
    mu=[]
    psi = []
    phi=[]
    pressure = []
    rho = []
    indepres = []
    res1 = []
    res2 = []
    res3 = []
    res4 = []


    datafile=open(filename,'r')
    lines = datafile.readlines()[1:]
    for line in lines:
        (a,b,c,d,e,f,g,g1,g2,g3,g4)=line.split()
        r.append(float(a))
        mu.append(float(b))
        psi.append(float(c))
        phi.append(float(d))
        pressure.append(float(e))
        rho.append(float(f))
        indepres.append(float(g))
        res1.append(float(g1))
        res2.append(float(g2))
        res3.append(float(g3))
        res4.append(float(g4))


    datafile.close()
    return np.array(r),np.array(mu),np.array(psi),np.array(phi),np.array(pressure),np.array(rho),np.array(indepres),np.array(res1),np.array(res2),np.array(res3),np.array(res4)
```

```python
numberOfFiles = 12
txt_files = sorted(glob.glob('star2/*.txt'), key=os.path.getmtime)
colors = cm.rainbow(np.linspace(0, 1.2, numberOfFiles))
h = 2 # (order of convergence )**
```

increasing number indicates increasing mesh points, starting from n=64 to n=131072. The following is the solution for beta = +200 (not code units) and mass_st = 0.

```python
###MU###
r_prev,mu_prev,_,_,_,_,_,_,_,_,_= read_data(txt_files[0])
plt.figure()
fig = plt.gcf()
fig.set_size_inches(9, 5)
plt.title("mu")
i=0
for filename in txt_files[1:]:
    _,mu,_,_,_,_,_,_,_,_,_ = read_data(filename)
    plt.semilogx(r_prev, (mu_prev[::2**i]-mu[::2**(i+1)])*h**(i),color=colors[i],label=i)
    mu_prev = mu
    i+=1
plt.grid()
plt.legend()
# plt.ylim(-3e-5,3e-5)
```

```python
##PSI####
r_prev,_,psi_prev,_,_,_,_,_,_,_,_= read_data(txt_files[0])
plt.figure()
fig = plt.gcf()
fig.set_size_inches(9, 5)
plt.title("psi")
i=0
for filename in txt_files[1:]:
    _,_,psi,_,_,_,_,_,_,_,_ = read_data(filename)
    plt.semilogx(r_prev, (psi_prev[::2**(i)]-psi[::2**(i+1)])*h**i,color=colors[i],label=i)
    psi_prev = psi
    i+=1
plt.grid()
plt.legend()
# plt.ylim(-7e-3, 7e-3)
```

```python
##PHI####
r_prev,_,_,phi_prev,_,_,_,_,_,_,_= read_data(txt_files[0])
plt.figure()
fig = plt.gcf()
fig.set_size_inches(9, 5)
plt.title("phi")
i=0
for filename in txt_files[1:]:
    _,_,_,phi,_,_,_,_,_,_,_ = read_data(filename)
    plt.semilogx(r_prev, (phi_prev[::2**(i)]-phi[::2**(i+1)])*h**i,color=colors[i],label=i)
    phi_prev = phi
    i+=1
plt.grid()
plt.legend(loc ='right')
# plt.ylim(-1e-4, 1e-4)
```

```python
##Pressure####
r_prev,_,_,_,pressure_prev,_,_,_,_,_,_= read_data(txt_files[0])
plt.figure()
fig = plt.gcf()
fig.set_size_inches(9, 5)
plt.title("pressure")
i=0
for filename in txt_files[1:]:
    _,_,_,_,pressure,_,_,_,_,_,_ = read_data(filename)
    plt.semilogx(r_prev, (pressure_prev[::2**(i)]-pressure[::2**(i+1)])*h**i,color=colors[i],label=i)
    pressure_prev = pressure
    i+=1
plt.grid()
plt.legend(loc="right")
# plt.ylim(-3e-1, 3e-1)
```

```python
##RHO##
r_prev,_,_,_,_,rho_prev,_,_,_,_,_= read_data(txt_files[0])
plt.figure()
fig = plt.gcf()
fig.set_size_inches(9, 5)
plt.title("rho")
i=0
for filename in txt_files[1:]:
    _,_,_,_,_,rho,_,_,_,_,_ = read_data(filename)
    plt.semilogx(r_prev, (rho_prev[::2**(i)]-rho[::2**(i+1)])*h**i,color=colors[i],label=i)
    rho_prev = rho
    i+=1
plt.grid()
plt.legend(loc='right')
# plt.ylim(-3e-2, 3e-2)
```
