```python
import numpy as np
import matplotlib.pyplot as plt
import glob
import os
from matplotlib import cm
```

```python
def read_data(filename):
    r=[]
    mu=[]
    psi = []
    phi=[]
    pressure = []
    rho = []


    datafile=open(filename,'r')
    lines = datafile.readlines()[1:-1]
    for line in lines:
        (a,b,c,d,e,f)=line.split()
        r.append(float(a))
        mu.append(float(b))
        psi.append(float(c))
        phi.append(float(d))
        pressure.append(float(e))
        rho.append(float(f))
    datafile.close()
    return np.array(r),np.array(mu),np.array(psi),np.array(phi),np.array(pressure),np.array(rho)
```

```python
numberOfFiles = 12
txt_files = sorted(glob.glob('*.txt'), key=os.path.getmtime)
colors = cm.rainbow(np.linspace(0, 1.2, numberOfFiles))
print("increasing number indicates increasing mesh points, starting from n=64 to n=131072")
exlude_last = 0 #stars
```

```python
###MU###
r_prev,mu_prev,psi_prev,phi_prev,pressure_prev,rho_prev= read_data(txt_files[0])
plt.figure()
fig = plt.gcf()
fig.set_size_inches(9, 5)
plt.title("mu")
i=0
for filename in txt_files[1:numberOfFiles-exlude_last]:
    r,mu,psi,phi,pressure,rho = read_data(filename)
    plt.semilogx(r_prev, (mu_prev-mu[::2])*2**(i),color=colors[i],label=i)
    mu_prev = mu
    r_prev = r
    i+=1
plt.grid()
plt.legend()
```
